```python
import pysrt
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read srt file (subtitles) and calculate CPS(characters per sceond) of subtitles
subs = pysrt.open('sample.srt', encoding='iso-8859-1')
```

```python
# Calcualte average CPS

final_data = np.zeros((3,len(subs)))
j=0
                      
for i in subs:
    tmp = subs[j]
    
    # Caculate the duaration of displaying one subtitle
    if tmp.end.minutes == tmp.start.minutes:
        final_data[1][j] = tmp.end.seconds - tmp.start.seconds
    else: 
        final_data[1][j] = (tmp.end.minutes * 60 + tmp.end.seconds) - (tmp.start.minutes * 60 + tmp.start.seconds)        
    
    # Clean subtitles
    sub_clean = sub_clean.replace("\\","")
    sub_clean = sub_clean.replace("\n","")     
    sub_clean = tmp.text.replace(" ","")
    sub_clean = sub_clean.replace(",","")
    sub_clean = sub_clean.replace(".","")
    cps_s = len(sub_clean) / final_data[1][j]
                      
    final_data[0][j] = cps_s
    j += 1
                      
len(final_data[1])

# Caculate an average cps of a video
cps_avg = sum(final_data[0]) / len(final_data[0])
cps_avg
```
